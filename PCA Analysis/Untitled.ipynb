### SCALING DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(pixels)
```

```python
scaled_X
```

### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_pixels=pca.fit_transform(pixels)
pca_pixels
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10,6),dpi=150)
labels = df['number_label'].values
sns.scatterplot(pca_pixels[:,0],pca_pixels[:,1], hue=labels, palette='Set1')
```

### 3D PLOT

```python
from sklearn.decomposition import PCA
pca_model = PCA(n_components=3)
pca_pixels = pca_model.fit_transform(scaled_X)
```

```python
from mpl_toolkits import mplot3d
```

```python
plt.figure(figsize=(8,8),dpi=150)
ax=plt.axes(projection='3d')
ax.scatter3D(pca_pixels[:,0],pca_pixels[:,1],pca_pixels[:,2],c=df['number_label']);
```
